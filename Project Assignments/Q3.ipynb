```python
# Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv("diabetes_dataset.csv")
```

Q3: How accurately does a person’s self-reported "General Health" match up with their actual BMI and Blood Pressure?

```python
sns.jointplot(data, x='bmi', y = 'physical_activity_minutes_per_week', kind='hex')
plt.title("BMI versus Physcial Activity Per Week", y= 1.21)
```

```python
sns.jointplot(data, x='bmi', y = 'screen_time_hours_per_day', kind='hex')
plt.title("BMI versus Screen Time Per Day", y= 1.21)
```

```python
sns.jointplot(data, x='bmi', y = 'alcohol_consumption_per_week', kind='hex')
plt.title("BMI versus Alcohol Consumption Per Week", y= 1.21)
```

```python
sns.jointplot(data, x='systolic_bp', y = 'physical_activity_minutes_per_week', kind='hex')
plt.title("Systolic Blood Pressure versus Physcial Activity Per Week", y= 1.21)
plt.xlabel("Systolic Blood Pressure")
```
