### Q4: Plot a regression line and confidence interval
Select two of the features from your dataset and create a lmplot that shows the linear regression between them. Filter your data as you see fit. Include a meaningful title for the plot.

```python
import seaborn as sns
```

```python
diabetes_sample_1k = diabetes_data.sample(1000)
```

**Positive Correlation:**
The `lmplot` reveals a clear positive linear trend, suggesting that higher body mass is associated with elevated fasting blood sugar levels.

**Statistical Confidence:** The inclusion of a confidence interval around the regression line provides a visual measure of the uncertainty in the estimate; the relatively narrow band suggests a consistent relationship across the 1,000 samples.

**Data Distribution:** The scatter plot underlying the regression line shows the density and spread of individual patient data points across various BMI categories.

```python
reg = sns.lmplot(x='bmi', y='glucose_fasting', data=diabetes_sample_1k, height=6, aspect=1.5)

reg.fig.suptitle('Regression Analysis: BMI vs Fasting Glucose', fontsize=16)
reg.fig.subplots_adjust(top=0.9)

plt.show()
```

### Q5. Create a relplot that shows multiple features in a single plot.
Select two features from your dataset and plot a scatter plot. Select two additional features you want to highlight in your data points using `hue` and `size`. Include a meaningful title to the plot.

```python
sns.relplot(x='bmi', y='glucose_fasting', data=diabetes_sample_1k, hue='gender', size='age', height=6, aspect=1.5, palette='pastel')
plt.title('Scatter Plot with Additional Features: BMI vs Fasting Glucose')
plt.show()
```
