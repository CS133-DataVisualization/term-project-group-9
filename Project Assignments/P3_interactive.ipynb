## Question 2. Create a simple interaction plot
Create a plot, it can be related to your question #1 or a new question, that users can interact with in some meaningful way. Explain in a markdown, how does the interaction add to the analysis.

Example of possible interactions:
*   Hover over information
*   Toogle between groups

```python
#It's hard to visualize the dataset with all 100k row so I kept it to only 20k sample, but I would reccomend changing the random state number
#each time you run if you want to see more than one sample of 20k rows
plot_sample = data.sample(n=20000, random_state=40)

fig_q2 = px.scatter(
    plot_sample,
    x="bmi",
    y="glucose_fasting",
    color="diabetes_stage",
    size="physical_activity_minutes_per_week",
    size_max=14,
    hover_name="gender",
    hover_data={
        "age": True,
        "diagnosed_diabetes": True,
        "hba1c": ":.2f",
        "bmi": ":.1f",
        "glucose_fasting": True,
        "physical_activity_minutes_per_week": True,
    },
    title="BMI vs fasting glucose by diabetes stage (bubble size = weekly activity minutes)",
    labels={
        "bmi": "BMI",
        "glucose_fasting": "Fasting glucose (mg/dL)",
        "diabetes_stage": "Diabetes stage",
        "physical_activity_minutes_per_week": "Activity (min/week)",
    },
    opacity=0.55,
)

fig_q2.update_traces(marker=dict(line=dict(width=0)))
fig_q2.update_layout(legend_title_text="Stage", hovermode="closest")

fig_q2.show()
```

**How this interaction adds to the analysis:**

The interaction allows us to be able to hover over each data point to shows the age, HbA1c, diagnosis, and activity level for each point so that we can identify and analyze outliers and clusters without overcrowding the chart. In addition, the legend on the right allows us to clicks on different types of diabetes stages to turn the stages on and off to see any overlapping or serparation between the different groups. The zoom interaction allows us to zoom into a smaller region for deeper inspection, especially in regions that are very dense with data points. The pan function allows us to move freely around the zoomed in area, which makes it easier to see a wider range of data points. In conclusion, with such a big dataset of 100k rows, even when we have limited the plot to only randomly include 20k of the 100k datas, the plot still shows very dense regions of datas. Therefore, without the ability to interact with the plot, it would be very difficult for us to explore the pattern between BMI vs fasting glucose by diatbetes stages.

When interacting with the scatter plot, I found that a majority of people with Type 2 diabetes has a fasting glucose level of around 80 to 160 mg/dL with most of the datapoints between 100-130 mg/dL. However, between the BMI level of ~23 and 30, there seems to be more data points clusters for people with Type 2 diabetes. ## Question 5. Geopy / Location Preparation

If your dataset has location names, addresses, cities, states, or countries, use geopy on a sample of your data to geocode locations or validate location information.

If your dataset does not have data that contains location, pick 5 destination you want to visit and use geopy to geocode locations.

Our data does not have locations, so I have picked Tokyo, Paris, Stockholm, Beijing, Sydney

```python
# Your code . . .
# Converted the city destinations into table with city, longitude, and latitude

destinations = {'City': ['Tokyo', 'Paris', 'Stockholm', 'Beijing', 'Sydney']}
geolocator = Nominatim(user_agent="CS133")
longitude = []
latitude = []
df = pd.DataFrame(destinations)
# df
x = {}
for location in destinations['City']:
    x[location] = geolocator.geocode(location)
    longitude.append(x[location].longitude)
    latitude.append(x[location].latitude)


df['longitude'] = longitude
df['latitude'] = latitude
df
```

## Question 6. Panel Widget

```python
# Your code . . .
# I made a selection and slider widget to see the mean diabetes risk score for each income group by education and weekly alcohol amount
data2 = data
alcohol_widget = pn.widgets.IntSlider(name ='Weekly Alcohol', start = data2['alcohol_consumption_per_week'].min(), end =data2['alcohol_consumption_per_week'].max())
unique_education = list(data['education_level'].unique())
edu_widget = pn.widgets.Select(name = 'Education', options = unique_education)
data2 = data2.interactive()

diabetes_pipeline = (
    data2[(data2['education_level']== edu_widget) & (data2['alcohol_consumption_per_week'] <= alcohol_widget)].groupby(['income_level'])['diabetes_risk_score'].mean().to_frame().reset_index()

)
diabetes_pipeline.hvplot.table()
```
